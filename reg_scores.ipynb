```python
import numpy as np
import pandas as pd
```

```python
dis = pd.read_csv(r"data\unique_disease_signatures.csv")
test_dis = pd.read_csv(r"data\testing_data.csv")['DOID']
drugs = pd.read_csv(r'data\unique_drugs.csv')
```

```python
dis=dis[dis['disease'].isin(test_dis)]
dis.head()
```

```python
drugs
```

```python

disease_gene_scores = {}
for _, row in dis.iterrows():
    disease = row['disease']
    genes = [
        g.replace("'", '').replace("[", '').replace("]", '')
        for g in row['up_genes'].split(' ')
    ]
    scores = [
        float(s.replace("'", '').replace("[", '').replace("]", ''))
        for s in row['up_reg_scores'].split(' ')
    ]
    disease_gene_scores[disease] = dict(zip(genes, scores))

drug_up_map = {}

for _, row in drugs.iterrows():
    if row['regulation_type'] != 'down':
        continue

    drug = row['drug_name']
    
    genes = [
        g.replace("'", '').replace("[", '').replace("]", '')
        for g in row['genes'].split(',')
    ]
    
    scores = [
        float(s.replace("'", '').replace("[", '').replace("]", ''))
        for s in row['scores'].split(',')
    ]
    
    for g, s in zip(genes, scores):
        drug_up_map.setdefault(g, []).append((drug, s))

down_reg_dict={}
for disease, gene_scores in disease_gene_scores.items():
    for gene, dis_score in gene_scores.items():
        down_reg_dict.setdefault(gene, {})[disease] = dis_score
        
        if gene in drug_up_map:
            for drug, drug_score in drug_up_map[gene]:
                down_reg_dict[gene][drug] = drug_score
```

```python
df_results = pd.DataFrame.from_dict(down_reg_dict, orient='index')
df_results.index.name = 'Gene' 
df_results.to_csv('disease_drug_interactions_down.csv')
```
