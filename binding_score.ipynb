```python
import pandas as pd
import time
import os

# ----------------------
# Load CSVs
# ----------------------
drug_df = pd.read_csv("/content/data/ddms1.csv")        # drug_name, smiles
protein_df = pd.read_csv("/content/data/disease_drug_interactions_with_sequences.csv")  # gene_name, amino_acid_seq

drug_names = drug_df["drug_name"].tolist()
smiles_list = drug_df["smiles"].tolist()

# If oneliner expects SMILES as X_repurpose
X_repurpose = smiles_list

# Matrix to store results
result_matrix = pd.DataFrame(
    index=protein_df["gene_names"],
    columns=drug_names
)

# ----------------------
# Loop over proteins
# ----------------------
for idx, row in protein_df.iterrows():
    gene_name = row["gene_names"]
    amino_seq = row["Amino Acid Sequence"]
    if not pd.isna(amino_seq):
      print(amino_seq)
      print(f"Running repurposing for {gene_name}...")

      start = time.time()

      # Build target from amino acid sequence
      # Adjust if your dataset loader is different
      target, target_name = amino_seq, gene_name

      os.makedirs(f'/content/save_folder/{gene_name}', exist_ok=True)
      scores = oneliner.repurpose(
          target=target,
          target_name=target_name,
          X_repurpose=X_repurpose,
          drug_names=drug_names,
          save_dir=f'/content/save_folder/{gene_name}',
          agg='mean'
      )

      end = time.time()
      print(f"Time lapse: {end - start:.2f}s")

      # ----------------------
      # Store results
      # ----------------------
      result_matrix.loc[gene_name] = scores

# ----------------------
# Save final matrix
# ----------------------
result_matrix.to_csv("drug_gene_matrix.csv")

print("Saved drug_gene_matrix.csv")
```

```python
def parse_binding_txt(file_path):
    records = []

    with open(file_path, "r") as f:
        for line in f:
            line = line.strip()

            # Must be a proper row
            if not line.startswith("|"):
                continue

            if "Rank" in line or "---" in line:
                continue

            parts = [p.strip() for p in line.split("|")[1:-1]]

            if len(parts) != 4:
                continue

            rank, drug, target, score = parts

            # Skip empty or invalid scores
            if score == "":
                continue

            try:
                score = float(score)
            except ValueError:
                continue

            records.append({
                "gene": target,
                "drug": drug,
                "score": score
            })

    return pd.DataFrame(records)
```

```python
import os
import glob
from pathlib import Path
import pandas as pd
txt_dir = list(Path("/content/save_folder").glob("**/*_aggregation/repurposing.txt"))
txt_dir
```

```python
for file in txt_dir:
    txt_path = str(file)
    df = parse_binding_txt(txt_path)

    for _, row in df.iterrows():
        gene = row["gene"]
        drug = row["drug"]
        score = row["score"]

        if gene in result_matrix.index and drug in result_matrix.columns:
            result_matrix.loc[gene, drug] = score
```
