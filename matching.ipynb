```python
import numpy as np

l = {}

def add_signature_data(gene, entity, score):
    # Clean gene name and ignore empty strings
    gene = gene.strip().replace("'", "").replace("[", "").replace("]", "").replace("\\n", "")
    if not gene: return
    if gene not in l:
        l[gene] = {}
    l[gene][entity] = score

# 1. Process Disease Signatures (including both up and down genes)
for _, row in dis.iterrows():
    d_name = row['disease']

    # Up genes
    up_g = row['up_genes'].strip("[]").replace("'", "").split()
    up_s = [float(x.replace("'", "")) for x in row['up_reg_scores'].strip("[]").replace("'", "").split()]
    for g, s in zip(up_g, up_s): add_signature_data(g, d_name, s)

    # Down genes
    dn_g = row['down_genes'].strip("[]").replace("'", "").split()
    dn_s = [float(x.replace("'", "")) for x in row['down_reg_scores'].strip("[]").replace("'", "").split()]
    for g, s in zip(dn_g, dn_s): add_signature_data(g, d_name, s)

# 2. Process Drug Signatures
for _, row in drugs.iterrows():
    dr_name = row['drug_name']
    genes = [g.strip() for g in str(row['genes']).split(',')]
    scores = [float(s.strip()) for s in str(row['scores']).split(',')]
    for g, s in zip(genes, scores): add_signature_data(g, dr_name, s)

df_results = pd.DataFrame.from_dict(l, orient='index')
print(f"Processed signature matrix with {df_results.shape[0]} genes and {df_results.shape[1]} total entities.")
```

```python
disease_cols = [col for col in df_results.columns if col.startswith('DOID:')]
drug_cols = [col for col in df_results.columns if col not in disease_cols]

# Initialize matching_scores_df with drugs as index and diseases as columns
matching_scores_df = pd.DataFrame(0.0, index=drug_cols, columns=disease_cols)

# Use negative product to score counteraction:
# - Opposite signs (reversal) -> Positive score contribution
# - Same signs (exacerbation) -> Negative score contribution
# - Missing data/No effect     -> Zero contribution

for disease_col in disease_cols:
    # Get genes relevant to this specific disease
    d_sig = df_results[disease_col].dropna()

    for drug_col in drug_cols:
        # Get drug scores for the genes involved in this disease
        dr_sig = df_results.loc[d_sig.index, drug_col].dropna()

        if not dr_sig.empty:
            # Calculate counteraction score: sum of -(disease_score * drug_score)
            # This rewards opposite directions and penalizes same directions
            score = -(df_results.loc[dr_sig.index, disease_col] * dr_sig).sum()
            matching_scores_df.loc[drug_col, disease_col] = score

print("Matching scores calculated using negative product logic.")
display(matching_scores_df.head())
```
